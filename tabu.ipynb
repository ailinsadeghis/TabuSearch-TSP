```python


def tabu_search(initial_tour, tabu_size, max_iter):
    current_tour = initial_tour
    best_tour = initial_tour
    tabu_list = []

    history = [current_tour]
    max_iters = max_iter


    with tqdm.tqdm(total=max_iters, leave=False , postfix={'current': "{:.2f}".format(current_tour.length()), 'percent': "{:.2f}%".format(0)}) as pbar:

        
        for _ in range(max_iter):
            current_best_neighbor = None
            current_best_length = float('inf')
            # local search + tabu list check
            for _ in range(100):  
                neighbor = current_tour.random_neighbor()
                neighbor_length = neighbor.length()
                if neighbor_length < current_best_length and neighbor not in tabu_list:
                    current_best_neighbor = neighbor
                    current_best_length = neighbor_length
                    
            
            if current_best_neighbor is None:
                break
            # move
            current_tour = current_best_neighbor
            if current_tour.length() < best_tour.length():
                best_tour = current_tour
                history.append(best_tour)
                
            # append and pop tabue list
            tabu_list.append(current_best_neighbor)
            if len(tabu_list) > tabu_size:
                tabu_list.pop(0)


            # process bar
            pbar.set_postfix({'current': "{:.2f}".format(current_tour.length())})
            pbar.update(1)
           
        pbar.close()

        
        return best_tour , history

```

```python
N = 10
tabu_size = 5
max_iter = 5
initial_tour = Tour(ir[:N])

start_time = time.time()

best_tour , his = tabu_search(initial_tour,tabu_size,max_iter)

end_time = time.time()
```
